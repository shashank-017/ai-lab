```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

text = "Obama was born on August 4 in Honolulu, and he loved using Apple products, he also loved apple in morning"
tokens = word_tokenize(text)

pos_tags = pos_tag(tokens)
named_entities = ne_chunk(pos_tags)

named_entities
```
