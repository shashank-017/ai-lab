```python
import nltk
nltk.download('punkt_tab')
```

```python
#Tokenisation based on words
```

```python
from nltk.tokenize import word_tokenize
text = "I am Shashank and I love to work on NLP!"
token = word_tokenize(text)
token
```

```python
#Tokenisation based on sentences
```

```python
from nltk.tokenize import sent_tokenize
text = "I am really loving NLP! I wish I can learn more and more."
sentences = sent_tokenize(text)
sentences
```

```python
#Tokenisation based on Regular Expression
```

```python
from nltk.tokenize import RegexpTokenizer
text = "I am really loving NLP! I wish I can learn more and more. It's so interesting"
tokenizer = RegexpTokenizer(r"\w+")
tokens = tokenizer.tokenize(text)
tokens
```

```python
#White Space Tokenizer
```

```python
from nltk.tokenize import WhitespaceTokenizer
tokenizer = WhitespaceTokenizer()
tokens = tokenizer.tokenize(text)
tokens
```

```python
#TweetTokenizer - specially made for Tweets ( it may contains special chars like @ or # and also emojis)
```

```python
from nltk.tokenize import TweetTokenizer
tokenizer = TweetTokenizer()
text = "Hey @shashank! I just discovered a new concept in NLTK!!!!! #Tech #LearningByDoing 🥷🦁"
tokens = tokenizer.tokenize(text)
tokens
```
