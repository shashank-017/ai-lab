```python
import nltk
```

```python
sentence = "Shashank is learning NLP"

tokens = nltk.word_tokenize(sentence)
pos_tagging = nltk.pos_tag(tokens)
#named_entities = nltk.ne_chunk(pos_tagging)
print(pos_tagging)
```

```python
named_entities = nltk.ne_chunk(pos_tagging)
named_entities
```

```python
sentence = "The boy visits the doctor every month. He is advised to eat apples"

tokens = nltk.word_tokenize(sentence)
pos_tagging = nltk.pos_tag(tokens)
print(pos_tagging)
```

```python
named_entities = nltk.ne_chunk(pos_tagging)
named_entities
```
