```python
import spacy

nlp = spacy.load('en_core_web_sm')  # en_core_web_sm this is a pre-train language model
```

```python
text = "the quick brown fox jumps over the lazy dog."
doc = nlp(text)

print(f'{'Token':<15}{'POS':<15}{'TAG':<15}{'Explaination'}')
print('-'*100)
for tokens in doc:
    print(f'{tokens.text:<15}{tokens.pos_:<15}{tokens.tag_:<15}{spacy.explain(tokens.tag_)}')
```
