```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = "The quick brown fox jumps over the lazy dog."
doc = nlp(text)

print(f'{'Token':<15}{'Head':<15}{'Dep':<15}{'POS':<15}{'Explanation'}')
print('-'*110)
for token in doc:
    print(f'{token.text:<15}{token.head.text:<15}{token.dep_:<15}{token.pos_:<15}{spacy.explain(token.tag_)}')
```
