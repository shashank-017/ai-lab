```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = "The children are running faster today than yesterday."
doc = nlp(text)

print(f"{'Token':<15}{'Lemma':<15}{'POS':<15}{'Explanation'}")
print('-'*60)

for token in doc:
    print(f"{token.text:<15}{token.lemma_:<15}{token.pos_:<15}{spacy.explain(token.pos_)}")
```
