```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = "Apple was founded by Steve Jobs and is headquartered in California, he might found it while eating Apple! and MS Dhoni loves using it"
doc = nlp(text)

print(f"{'Entity Text':<25}{'Label':<15}{'Explanation'}")
print('-'*80)

for ent in doc.ents:
    print(f"{ent.text:<25}{ent.label_:<15}{spacy.explain(ent.label_)}")
```
